```python
import pandas as pd
import json
```

## 1. Create a single dataframe with the concatenation of all input csv files, adding a column called country

```python
countries = [
    "US", "RU", "JP", "KR", "MX" ,"CA", "DE", "FR", "GB", "IN"
]

frames = []
for country in countries:
    df = pd.read_csv('data/' + country + "videos.csv.zst", encoding="utf-8", encoding_errors="ignore")
    df["country"] = country
    frames.append(df)

all_videos = pd.concat(frames, ignore_index=True)
all_videos.head()
```

## 2. Extract all videos that have no tag.

```python
all_videos['tags'] = all_videos['tags'].apply(lambda tags_string: [] if tags_string == '[none]' else tags_string.split('|'))
all_videos['tags'] = all_videos['tags'].apply(lambda tags: [tag.replace('"','') for tag in tags])

all_videos['tags']
```

```python
# all_videos['tags'] == ['none']

no_tags = all_videos[all_videos['tags'].apply(lambda tags: len(tags) == 0)]
no_tags
```

## 4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
mask = (
    (unique_df['comments_disabled'] == True) &
    (unique_df['ratings_disabled'] == True) |
    (unique_df['video_error_or_removed'] == True)
)

# just in case we need the original dataset later
org_df = unique_df.copy()

excluded = unique_df[mask]
unique_df = unique_df[~mask]
excluded
```

## 5. Add a like_ratio column storing the ratio between the number of likes and of dislikes

```python
dislikes = unique_df['dislikes'].replace(0, pd.NA)
likes = unique_df['likes'].replace(0, pd.NA)

unique_df['like_ratio'] = likes / dislikes
unique_df[['likes', 'dislikes', 'like_ratio']]
```

## 6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
publish_dt = pd.to_datetime(unique_df['publish_time'], errors='coerce', utc=True)

# create a series of date times
# convert the date time to the nearest 10 minutes
interval_start = publish_dt.dt.floor('10min')

# add 10 minutes
interval_end = interval_start + pd.Timedelta(minutes=10)

unique_df['publish_interval_10min'] = (
    interval_start.dt.strftime('%H:%M') + ' - ' + interval_end.dt.strftime('%H:%M')
)

unique_df[['publish_time', 'publish_interval_10min']].head()
```

## 7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
output_df = unique_df.groupby('publish_interval_10min')['video_id'].size().to_frame(name='video_count')

output_df['dislike_avg'] = unique_df.groupby('publish_interval_10min')['dislikes'].mean().round()
output_df['like_avg'] = unique_df.groupby('publish_interval_10min')['likes'].mean().round()

output_df
```

## 14. Read all json files with the video categories

```python
countries = [
    "US", "RU", "JP", "KR", "MX" ,"CA", "DE", "FR", "GB", "IN"
]

frames = []
for country in countries:
    json_name = 'data/' + country + "_category_id.json"
    data = pd.read_json(json_name)
    df = pd.json_normalize(data['items'])
    df["country"] = country
    frames.append(df)

categories = pd.concat(frames, ignore_index=True)
categories
```

## 15. For each country, determine how many videos have a category that is not assignable.
